```python
import pandas as pd
```

```python
df = pd.read_csv("House Price India.csv")
df.head(3)
```

```python
df.info()
```

```python
df.describe()
```

```python
stats = df['Price'].describe().reset_index()
stats['Price'] = round(stats['Price'],2)
```

```python
stats
```

```python
df.isna().sum().sum()
```

```python
df.duplicated().sum()
```

```python
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
```

```python
import matplotlib.pyplot as plt
```

```python
df.groupby("condition of the house")['Price'].mean().sort_values(ascending=False).plot(kind="bar")
plt.title("Condition of the House vs Price")
plt.xlabel('Condition of the House')
plt.ylabel("Mean Price")
plt.show()
```

```python
df.columns
```

```python
X = df[['number of bedrooms', 'number of bathrooms','living area','condition of the house','Number of schools nearby']]
y = df[['Price']]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.tree import DecisionTreeRegressor
param_grid = {
    'criterion' : ['mse', 'friedman_mse', 'mae'],
    'splitter' : ['best','random'],
    'max_depth' : [None, 10, 20, 30, 40, 50],
    'min_samples_split' : [2,5,10],
    'min_samples_leaf' : [1,2,4]
    }
```

```python
tree_model = DecisionTreeRegressor()
```

```python
grid_tree =GridSearchCV(estimator = tree_model, param_grid = param_grid )
```

```python
import warnings
warnings.filterwarnings("ignore", category=UserWarning)
from sklearn.exceptions import FitFailedWarning
import warnings
warnings.simplefilter("ignore", FitFailedWarning)
```

```python
grid_tree.fit(X_train, y_train)
```

```python
grid_tree.best_params_
```

```python
tree_preds = grid_tree.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, tree_preds)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
predslr = lr.predict(X_test)
```

```python
mean_absolute_error(y_test, predslr)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfmodel = RandomForestRegressor() 
```

```python
param_gridfr = {
    'max_depth' : [5,10,15],
    "n_estimators" : [2,3,4,5,6,7,8,9,10]
}
```

```python
gridfr = GridSearchCV(rfmodel, param_gridfr)
```

```python
gridfr.fit(X_train, y_train.values.ravel())
```

```python
gridfr.best_params_
```

```python
rfrpredictions = gridfr.predict(X_test)
```

```python
mean_absolute_error(y_test,rfrpredictions)
```

```python
import joblib
joblib.dump(gridfr,"model.pkl")
```
